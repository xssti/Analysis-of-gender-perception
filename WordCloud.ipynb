```python
text = ['Text']
df = pd.read_csv("/content/drive/MyDrive/ADS-1/1900/Wordcloud_1928.csv",usecols= text)
```

```python
df.isna().sum()
```

```python
df['Text'].fillna("No Text", inplace = True)
```

```python
text = " ".join(review for review in df.Text.astype(str))
```

```python
print ("There are {} words in the combination of all cells in column Text Column.".format(len(text)))
```

```python
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
stopwords = set(STOPWORDS)
```

```python
stopwords.update(["think", "said", "going", "far", "now","will", 'really', "hon", "say","way","want","ask","good","regard","even"])
```

```python
wordcloud = WordCloud(stopwords=stopwords, background_color="black", width=1600, height=800).generate(text)
```

```python
plt.axis("off")
plt.figure(figsize=(40,20))
plt.tight_layout(pad=0)
plt.imshow(wordcloud, interpolation='bilinear')
plt.show()
```
